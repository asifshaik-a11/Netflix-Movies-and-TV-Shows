```python
#data cleaning
```

```python
#1. Rename Column Headers to be clean and uniform
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df
```

```python
# 2. Remove Duplicate Rows
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
# 3. Identify and Handle Missing Values
# Impute missing 'director', 'cast', and 'country' values with 'Unknown'
for col in ['director', 'cast', 'country']:
    df[col].fillna('Unknown', inplace=True)
```

```python
df
```

```python
# Drop rows where 'date_added' or 'duration' are missing (critical missing data)
df.dropna(subset=['date_added', 'duration'], inplace=True)
```

```python
df
```

```python
# Impute missing 'rating' with the mode (most frequent value: 'TV-MA')
mode_rating = df['rating'].mode()[0]
df['rating'].fillna(mode_rating, inplace=True)
```

```python
df
```

```python
# 4. Convert date formats to a consistent type (dd-mm-yyyy) and Check/Fix data types
# Convert 'date_added' to datetime, then format to 'dd-mm-yyyy' string
df['date_added'] = pd.to_datetime(df['date_added'])
df['date_added_formatted'] = df['date_added'].dt.strftime('%d-%m-%Y')
```

```python
df
```

```python
# Drop the old datetime column and rename the new formatted one
df.drop('date_added', axis=1, inplace=True)
df.rename(columns={'date_added_formatted': 'date_added'}, inplace=True)
```

```python
df
```

```python
# Reorder columns for better presentation (optional)
cols = ['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added', 
        'release_year', 'rating', 'duration', 'listed_in', 'description']
df = df[cols]
```

```python
# 5. Save the cleaned dataset
cleaned_file_path = "netflix_titles_cleaned.csv"
df.to_csv(cleaned_file_path, index=False)
```
